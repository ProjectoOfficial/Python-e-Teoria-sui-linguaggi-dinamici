# <span style="color:red">Differenze con Java</span>

Python si differenzia da Java su molteplici aspetti, tra i quali:

* FILES: In <span style="color:blue">Java</span> un file può contenere una e una sola classe; se la classe è pubblica, il nome della classe deve coincidere con il nome del file. Altre eventuali classi devono essere private. In <span style="color:blue">Python</span> un file può contenere 0, 1 o n classi.

* ESECUZIONE DI UN PROGRAMMA:<span style="color:blue">Java</span> richiede che una classe contenga il metodo pubblico e statico main(). In <span style="color:blue">Python</span> il codice viene eseguito sequenzialmente dall'inizio alla fine, non esiste una funzione main().

#### Type checking dinamico:

Il type checking è dinamico se la maggior parte delle operazioni (non per forza tutte) di type checking vengono effettuate a tempo di esecuzione. Questo implica che non è necessario esplicitare il tipo di una variabile, ci penserà l’interprete a tempo di esecuzione ad associare alla variabile il tipo corretto in base al dato che le stiamo assegnando.

Questa dinamicità permette più flessibilità rispetto al modello statico, con maggior overhead (più lavoro da parte dell’interprete nel gestire le risorse)
   * le variabili possono cambiare tipo a run-time
   * gestione della struttura dei dati a run-time (problemi di gestione della memoria, processi, tempo di esecuzione … )
   
Nonostante ciò, tutta questa flessibilità ha un prezzo. Un type checker dinamico dà minori garanzie a priori perché opera (la maggior parte del tempo) a run-time, quindi è molto possibile incorrere in tante situazioni di errore a causa di input inaspettati. Sarà quindi necessario:
   * gestire le situazioni di errore a runtime con un meccanismo di gestione delle eccezioni (try: / except: … import traceback as t, t.print_stack())
   * verificare la correttezza di un programma effettuando tanti test
   


# <span style="color:red">Esecuzione da linea di comando</span>
Il seguente script può essere utilizzato come template per crearne altri in grado di poter supportare la codifica "UTF-8". Stiamo parlando di ambiente linux!

````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from math import sqrt
def eq2grado(a,b,c):
    ''' inserire in input i coefficienti dell'equazione di secondo grado ax^2+by+c
    a -> non nullo
    output: una coppia (x1,x2) di radici dell'equazione. Nel caso in cui il delta sia negativo, 
    allora in output avrò due coppie composte da parte reale e parte immaginaria di due numeri complessi
    '''
    assert a != 0
    if b != 0:
        if c == 0:
            return -b/(2*a), -b/(2*a)
        else:
            d = b**2 -4*a*c
            if d>0:
                s = sqrt(d)
                return (-b-s)/(2*a),(-b+s)/(2*a)
            elif d == 0:
                return -b/(2*a), -b/(2*a) #inutile, controlliamo gia c
            else:
                s = sqrt(d)
                return (-b/(2*a), -abs(s/(2*a))), (-b/(2*a),abs(s/(2*a))
    else:
        if c == 0:
                return 0,0
        elif c > 0:
                return (0, -sqrt(abs(c/a))),(0,sqrt(abs(c/a)))
        else:
            return -sqrt(abs(c/a)),sqrt(abs(c/a))
                                                   
if __name__ == '__main__':
    a,b,c = [float(c) for c in input('inserire coefficienti a,b,c')]
    x1, x2 = eq2grado(a,b,c)
    print("soluzione 1: ",x1," - soluzione 2: ",x2)
````

Python rende esplicite le nozioni di oggetto <span style="font-style: italic; color:blue">iterabile</span> e di <span style="font-style: italic; color:blue">iteratore</span>

Un <span style="font-style: italic; color:blue">iterabile</span> è una struttura sulla quale è possibile <span style="font-style: italic;">iterare</span>, ovvero accedere ordinatamente a tutti gli elementi. Il costrutto <span style="font-style: italic; ">for</span> è un implementazione di un iteratore; nonostante ciò, non è l'unica.


ma questa soluzione non è <span style="font-style:italic">pythonic</span>. Sfruttando al meglio le possibilità che Python ci fornisce, possiamo scrivere:

```python
from random import random
```

Questa costruzione si chiama <span style="font-style: italic; color:blue">list comprehension</span>.



````python
#verifichiamo che range(10) sia un iterabile

from collections.abc import Iterable
if isinstance(range(10),Iterable):
    print("iterabile")
else:
    print("non iterabile")
    
# oppure possiamo controllare che abbia il metodo __iter__
try:
    hasattr(1,'_iter__')
    print("iterabile")
except TypeError:
    print("non iterabile")
````

````python
n = int(input())
N = range(n)
I = iter(N)
while True:
    try:
        print(next(I))
    except StopIteration:
        break

for i in range(n):
    print(i)
````

### Sintassi

Per rendere più semplice l'utilizzo dell'iteratore, Python provvede aggiungendo un po' di zucchero sintattico, è possibile scrivere:
```python
I = iter(T)
```
invece di
```python
I = T.__iter__()
```
analogamente, se I è un iteratore:

```python
E = next(I)
```
invece di 
```python
E = I.__next__()
```

Nel caso in cui stessimo iterando manualmente, ad esempio all'interno di un while, è nostro compito terminare l'iterazione quando saranno finiti gli elementi iterabili. Per terminare l'iterazione bisogna controllare il generarsi dell'eccezione <span style="font-style:italic">StopIteration</span>:

````python
n = int(input("inserire numero iterazioni: "))
I = iter(range(n))
while True:
    try:
        print(next(I))
    except StopIteration:
        break
````

Il costrutto <span style="color:blue">for</span> si occupa automaticamente di gestire questo procedimento (aggiungendo zucchero sintattico). L'esempio precedente è più che altro utile per capire cosa sta a dietro all'iterazione di un iterabile.

L'implementazione con il ciclo for è la seguente:

````python
n = int(input("inserire numero iterazioni: "))
for i in range(n):
    print(i)
````

````python
# Stesso esempio con un differente iterabile
# Prima la soluzione, elegante e con l'iteratore implicito
for c in ['a', 'b', 'c', 'd']:
    print(c)
# Poi quella con iteratore esplicito
L = ['a', 'b', 'c', 'd'] # La lista è un iterabile, quindi ...
I = iter(L)
while True:
    try:
        print(next(I))
    except StopIteration:
        break
````

## <span style="color:red">Classi e oggetti</span>

Molto più semplici rispetto a Java in quanto non esistono costrutti come <span style="font-style:italic"> public, protected, private, ecc.</span>
Gli obiettivi di Java sono più che altro garantire l'integrità del codice, prevenire gli errori ... L'obiettivo di Python è rendere più semplice la scrittura del codice.

In Python 3 tutti i tipi, incluse le primitive come <span style="color:blue">int</span> e <span style="color:blue">bool</span>, sono classi e <u>ogni oggetto è un'istanza di una classe</u>.

In generale, la filosofia di Python (e la conseguente implementazione) è che tutti gli oggetti sono <span style="font-style: italic; color:blue">first-class object</span>, ovvero ogni oggetto del linguaggio deve avere lo stesso stato e ognuno di essi deve poter essere inserito in una variabile, in una lista, in un dizionario, e così via, senza alcun problema.

## Differenze importanti

In Python non esistono le distizioni fatte da Java sullo *scope* di classi, metodi e attributi (public, private, protected ...).
<u> Tutto è pubblico</u>, anche se esistono nomenclature da utilizzare per differenziare l'uso di un elemento del programma.

In Python gli attributi di una classe vanno definiti all'interno della funzione \_\_init\_\_. Questa è soltanto una convensione, potrebbero essere definiti anche altrove, ma sarebbe un cattivo uso del linguaggio e fonte di errori.

Un osservazione fondamentale si basa sul fatto che gli attributi di un oggetto <u>*non possono essere dichiarati fuori dalle funzioni di una classe*</u>.
Gli attributi dichiarati al di fuori di un "metodo" della classe coincidono alle *variabili statiche* in Java. Per accedere alle variabili statiche è necessario utilizzare una sintassi diversa, altrimenti l'interprete va a cercare una variabile non statica con quel nome all'interno della classe e non la trova -> è necessario fare riferimento alla classe.

I metodi definiti al'interno di una classe e che accedono agli attributi di quella stessa classe, hanno sempre un primo parametro (convenzionalmente indicato con <span style="color:blue">self</span>), che durante la chiamata (al metodo) viene legato all'istanza dell'oggetto in questione. <span style="color:blue">Self</span> corrisponde al <span style="color:blue">This</span> di Java.
Il parametro self non dobbiamo passarlo in modo esplicito quando creo l'oggetto, automaticamente Python associa a self l'istanza dell'oggetto attuale. 

Quando l'oggetto viene creato non viene chiamata subit la funzione \_\_init\_\_ ma la funzione \_\_new\_\_ che alloca la memoria base su cui allocare poi dinamicamente le variabili (la new non va quasi mai ridefinita!). Successivamente la \_\_new\_\_ chiama la \_\_init\_\_ con il parametro self. La memoria dedicata alle variabili definite nella \_\_init\_\_ verrà allocata dinamicamente.

\_\_doc_\_ è un class attribute con un significato speciale (il suo valore è la docstring). Contiene la descrizione dell'intera classe.

## Getter e Setter

Un __getter__ è un metodo che restituisce il valore di un attributo

Un __setter__ è un metodo che assegna un valore ad un attributo

````python
class CC:
    """
    Classe per la gestione di un conto corrente: v2.0 
    """
    def __init__(self, deposito_iniziale = 0):
        print("Apertura conto con {0:.2f} Euro".format(deposito_iniziale))
        self.saldo = deposito_iniziale
        
    def prelievo(self, importo=0):
        # Se importo==0 equivale ad una lettura del saldo
        # Per questa ragione prelievo (con importo == 0)
        # può essere usato come "getter"
        if importo>0 and self.__saldo-importo>=0:
            self.__saldo -= importo
        return self.__saldo
    
    def __modifica_saldo(self, importo):
        # metodo usato come "getter"; forza il vincolo
        # sulla non negatività dell'importo
        if importo>=0:
            self.__saldo = importo
      
    def versamento(self, importo):
        # versamento non può essere un "setter" perché
        # modifica il valore sulla base di quello attuale
        if importo>0:
            self.__saldo += importo
     
    def mostra_saldo(self):
        print("Il saldo è di {0:.2f} Euro".format(self.__saldo))
    
    saldo=property(prelievo, __modifica_saldo) # getter e setter
````

# <span style="color:red">I Namespace</span>

* In ogni punto del programma, un nome può corrispondere a una variaile:
    
    1) <span style="font-style: italic; color:blue">locale</span>
    
    2) <span style="font-style: italic; color:blue">non locale</span>
    
    3) <span style="font-style: italic; color:blue">globale</span>
    
    4) <span style="font-style: italic; color:blue">built-in</span>
    
    5) oppure può non essere definito
    
In ogni punto del programma, i nomi locali, globali e predefiniti formano altrettante "collezioni" chiamate <span style="font-style: italic; color:blue">namespace</span>. In Python i namespace sono essi stessi manipolabili; sono <u>dizionari</u> e come tali possono essere consultati.

Quando ad esempio viene richiamata una funzione, cambia il namespace locale ma non quello globale; viene creato un nuovo dizionario con i nomi delle nuove variabili, ovvero quelle definite all'interno della funzione con l'aggiunta dei parametri. La creazione è dinamica e man mano che l'interprete trova nuove variabili, le aggiunge al namespace locale.

Per capire meglio i namespace non locali facciamo un esempio:
```python
    def B(x):
        const = 10
        return x + const
    def A(y):
        y = 0
        y = B(y)
        print(y)   
```
I nomi delle variabili di A per la funzione B sono in un namespace *non locale*.
        

Le funzioni *locals()* e *globals()* restituiscono rispettivamente i dizionari che contengono i nomi definiti nel namespace locale e i nomi definiti nel namespace globale.

Quando va in esecuzione l'interprete, l'ambiente locale e quello globale coincidono.
    

````python
locals() == globals()
````

Ovviamente i namespace contenuti all'interno di una funzione non corrispondono a quelli globali, quindi se dentro a una funzione scrivo *locals() == globals()* il risultato sarà *false*.

````python
def power2(p):
    powered = 2**p
    print(locals()==globals()) #sono diversi
    print(locals())# vediamo cosa contiene locals()
    
    #potete provare a scrivere print(globals()) 
    
    return powered
print(power2(10))
print(locals()==globals())
````

### Attenzione con le import

Quando si vuole importare una libreria si può procedere in diversi modi, dipende se si vuole importare una sola funzione o più funzioni contenute nella libreria.
```python
    #importo la sola funzione sqrt da math
    import math
    
    #in questo caso questa è la chiamata corretta
    x = math.sqrt(121)
    
    x = sqrt(121) # dà errore perche sqrt non è nel namespace locale e globale
```

In questo caso con *import math* andiamo ad importare la libreria math, e sarà lei a comparire nel nostro namespace globale e locale. Le funzioni presenti nella libreria math non vengono aggiunte ai namespace e quindi sarà necessario effettuare una chiamata alla funzione come mostrato nell'esempio corretto sopra.

```python
    #importo soltanto la funzione sqrt da math
    from math import sqrt
    
    #dà errore perchè math non è nei namespace
    x = math.sqrt(121)
    
    x = sqrt(121) # questa è la chiamata corretta
```

Ora vediamo che nel namespace locale e globale è stato aggiunto soltanto la funzione sqrt. Se scriviamo *print(math.sqrt(121)* il risultato non sarà *11* ma ***NameError: name 'math' is not defined***, in quanto math non è presente nè nel namespace locale nè nel globale.

```python
    #importo tutte le funzioni da math
    from math import *
    
    #dà errore per il motivo precedente
    x = math.sqrt(121)
    
    #chiamata corretta
    x = sqrt(121)
```

Questo utilizzo della import è quasi identico la precedente in quanto importa le funzioni di math, ma in questo caso le importa **tutte**. Bisogna sempre evitare di importare tutte le funzioni da una libreria in questo modo in quanto, se sono presenti nomi di funzioni che abbiamo dichiarato nel nostro script, verrà perso il riferimento al namespace globale e locale ad esse e sarà rimpiazzato da quello delle funzioni importate.


## <span style="color:red">Regole di scope</span>

Innanzitutto lo **scope** di una (dichiarazione di) variabile rappresenta tutti i punti del programma in cui tale variabile può essere acceduta. In poche parole *scope = visibilità di una variabile*.

L'interprete di Python usa la **regola LEGB** per determinare qual è la variabile (e quindi il valore) a cui si riferisce un dato nome:

   1) Prima il **namespace locale (L)**
   
   2) Poi i namespace locale alle **enclosing function (E)**
   
   3) Poi il **namespace globale(G)**
   
   4) Infine il namespace delle funzioni **built-in (B)**
   
*enclosing function =  funzioni innestate*

````python
def esterna(x):
    a = 5
    b = x
    print("locals esterna: ",locals())
    def interna():
        global a
        nonlocal b
        x = a+b
        a=2
        b=5
        return x
    a = interna()
    print("x: ",x)
    print("a: ",a)
    print("b: ",b)
    y = 3
    return abs(a+b)
    
````

Gli attributi di oggetti che rappresentano tipi predefiniti (ad esempio gli attributi degli oggetti che rappresentano le primitive) sono *Read-Only*, quindi non si possono aggiungere attributi nuovi a queste classi


````python
def myadd(L,M):
    ''' restituisce la concatenazione degli elementi di L
    con gli elementi di M in ordine inverso'''
    R = L[:] #copio tutta L in R
    T = M[:] #copio tutta M in T
    T.reverse()
    for t in T:
        R.append(t)
    return R
````

````python
#vediamo che non possiamo aggiungere la myadd alla classe list
setattr(list,'__add__',myadd) 
````

````python
#non possiamo farlo nemmeno ad un istanza della classe
L = [1,2]
setattr(L,'__add__',myadd)
````

### come aggiungere attributi a classi Read-Only

Se proprio si ha la necessità di modificare le proprietà di un oggetto predefinito, l'unico modo per aggiungere attributi e metodi a queste classi è creare una sottoclasse della classe read-only che definisca al suo interno questi metodi e attributi

````python
#definizione di classe derivata
class mylist(list):
    pass #copio tutti i metodi e attributi di list

def myadd(self,M):
    '''restituisce la concatenazione di self con
    gli elementi di M in ordine inverso'''
    
    R = self[:] #copio la lista self in R
    T = M[:] #copio la lista M in T
    T.reverse() #inverto l'ordine della lista T
    for t in T:
        R.append(t)
    return(mylist(R))

setattr(mylist,'__add__',myadd)
````

````python
L = mylist([1,2,3])
M = mylist([4,5,6])
print(L.__add__(M))
print(L)
````

````python
class graph:
    '''Versione 1.0: ciò che caratterizza un grafo è la definizione della sottoclasse edge
       per la rappresentazione degli archi. Un grafo coinciderà infatti, essenzialmente, con
       l'insieme dei suoi archi
    '''
    class edge:
        '''La classe edge definisce un arco pesato. L'arco è rappresentato come "tupla ordinata"
           degli estremi dell'arco stesso. Tupla e peso sono memorizzati come attributi.
        '''
        def __init__(self,u,v,weight=0):
            '''Se non specificato, il peso è considerato 0. Il grafo
            non è ordinato e dunque una tupla (x,y) 
            coincide con la tupla (y,x). Per semplificare
               altre operazioni, per l'arco (x,y) memorizzato vale x<y
            '''
            self.e = (min(u,v),max(u,v))  # Si noti che, con questa "soluzione" l'arco vero e proprio
            self.w = weight               # (cioè la coppia di nodi) e il peso sono posti allo "stesso livello"
        
        def __lt__(self,other):
            '''Un arco e1 è minore di un arco e2 se il peso di e1 è minore oppure se i pesi sono
               uguali ma e1 precede lessicograficamente e2
            '''
            return self.w<other.w or self.w==other.w and self.e<other.e
        
        def __eq__(self,other):
            '''Due archi sono uguali se hanno gli stessi estremi. Non somo cioè ammessi archi con
               stessi estremi ma pesi diversi
            '''
            return self.e[:2]==other.e[:2]
        
        def __hash__(self):
            '''Implementare questa funzione è necessario affinché 
            una classe "user-defined" rappresenti
               oggetti immutabili. Senza di essa non sarebbe 
               possibile costruire insiemi di archi'''
            return hash(self.e)
        
        def __str__(self):
            '''Restituisce una "rappresentazione esterna" dell'arco. Questa funzione è usata
               automaticamente da print().
            '''
            return str((self.e,self.w))
        
    def __init__(self,*args):
        '''Un grafo può essere creato specificando un numero arbitrario di tuple (coppie o terne)
           costituite dai due estremi che definiscono l'arco e l'eventuale peso
           Nodi e archi sono memorizzati in strutture dati set (insieme)
        '''
        self.edges = set()
        self.nodes = set()
        for e in args:
            self.edges.add(self.edge(*e))
            self.nodes.add(e[0])
            self.nodes.add(e[1])
            
    def __str__(self):
        '''Restituisce una "rappresentazione esterna" del grafo.'''
        return str([(edge.e,edge.w) for edge in self.edges])
        
    def sorted_edges(self):
        '''Restituisce la lista ordinata di archi'''
        return(sorted(self.edges))
    
    def add_edge(self,u,v,weight=0):
        '''Aggiunge, se non già presente, 
        l'arco (u,v) al grafo.
           Aggiorna l'insieme dei vertici,
           inserendo anche eventuali vertici isolati.
        '''
        self.edges.add(self.edge(u,v,weight))
        self.nodes = self.nodes.union({u,v})
        for x in range(1, max(u,v)+1):
            self.nodes.add(x)
        
````

````python
class graph:
    '''Versione 2.0: include una nuova definizione per 
    gli archi ma il resto dei metodi è immutato
       (come funzionalità
    '''
    class edge(tuple):
        '''edge ora è sottoclasse di tuple. Un edge è dunque una tupla con in più
           l'attributo che memorizza il peso e, come tale, può essere manipolato in modo
           standard proprio come una tupla
        '''
        def __new__(cls, u,v,weight=0):
            '''Per la creazione ridefiniamo il costruttore 
            __new__ della supeclasse
               passando i due estremi in ordine (prima il minore). Se si ridefinisce __new__
               non c'è chiamata automatica a init. L'eventuale chiamata deve essere
               esplicitamente fatta dentro __new__ ma ciò, quasi sempre, è inutile. Possiamo
               infatti mettere le altre inizializzazioni dentro __new__stessa. Nel nostro caso,
               si tratta solo di definire l'attributo che rappresenta il peso
            '''
            self = super().__new__(cls, (min(u,v), max(u,v)))
            self.w = weight
            return self
        
        def __lt__(self,other):
            '''Prestare qui attenzione alla differenza rispetto
            alla versione 1.0.
               Siccome gli archi sono tuple 
               (e la definizione "built-in" di tupla supporta
               il confronto), possiamo direttamente confrontare
               due archi anziché confrontare i loro
               (non più necessari) attributi "e" 
               (come nella versione 1.0). Non possiamo però usare
               l'operatore < perché questo produrrebbe
               una ricorsione infinita. Dobbiamo usare
               il confronto definito nella superclasse.
            '''
            return self.w<other.w or self.w==other.w and super().__lt__(other)
        
    def __init__(self, *args):
        '''Identica alla versione 1.0'''
        self.__edges = set()
        self.__nodes = set()   
        for e in args:
            self.add_edge(*e)
            
    def __str__(self):
        '''Notare la differenza con la 1.0'''
        return str(set([(edge,edge.w) for edge in self.edges]))
    
    def __edges(self):
        '''Identica alla 1.0'''
        return(sorted(self.__edges))
        
    def adj_list(self,u):
        '''Returns the sorted list of nodes adjacent to u in self'''
        return sorted([e[1] if e[0]==u else e[0] for e in self.edges if u in e])        
          
    def add_edge(self,u,v,weight=0):
        '''Aggiunge, se non già presente, l'arco (u,v) al grafo.
           Aggiorna l'insieme dei vertici, inserendo anche eventuali vertici isolati.
        '''
        self.__edges.add(self.edge(u,v,weight))
        self.__nodes = self.__nodes.union({u,v})
        for x in range(1, max(u,v)+1):
            self.__nodes.add(x)
            
    def __add_node(self,u):
        self.__nodes.add(u)
            
    def remove_edge(self,u,v):
        u,v = min(u,v),max(u,v)
        if v in self.adj_list(u):
            self.__edges.remove((u,v))
            
    def update_graph(self,edge_list):
        '''Called upon a direct assignment to the edge list of the graph.
           It simply create a new graph and copy nodes ad edges to self.
           This way, as long as the value assigned is a legal edge list 
           (possibly with weight) the modified graph is a legal graph.
        '''
        from copy import copy
        g = graph(*edge_list)
        self.__edges=copy(g.__edges)
        self.__nodes=copy(g.__nodes)
            
    edges=property(__edges, update_graph)
    nodes=property(lambda self: self.__nodes) # nodes can only be read
````

$\large u_1\quad v_1\quad w_1$

$\large u_2\quad v_2\quad w_2$

$\large\ldots$

$\large u_m\quad v_m\quad w_m$

il numero di archi del grafo coincide con il numero di terne (distinte nei primi due elementi che ne rappresentano gli estremi) nel file. il numero di vertici coincide invece con l'insieme degli elementi distinti presenti fra l'insieme degli estremi. Eventuali nodi isolati non vengono rappresentati

ad esempio, il grafo:

$\large 2\quad  3\quad  2$

$\large 5\quad  2\quad  4$

$\large 3\quad  4\quad  1$

$\large 5\quad  1\quad  2$

$\large 1\quad  3\quad  4$

$\large 7\quad  2\quad  4$

ha 6 vertici e i seguenti archi: (2,3), (5,2), (3,4), (5,1), (1,3), (7,2). Il vertice 6 è isolato. in questo caso inoltre, tutti i pesi sono interi positivi.

````python
def readgraph(fn):
    G = graph()
    with open(fn) as f:
        for l in f:
            tokens = l.strip().split(' ')
            
            u = int(tokens[0])
            v = int(tokens[1])
            w = int(tokens[2])
            G.add_edge(u,v,w)
    return G
````

Ovviamente non è che per ogni funzione possiamo scrivere prendere il tempo di inizio e il tempo di fine, ogni volta, per calcolare il tempo di esecuzione. Per risparmiare tempo e linee di codice, in Python, possiamo usare un decorator:

````python
def my_timer(f):
    from time import time
    def timedfun(n):
        start_time = int(round(time() * 1000))
        res = f(n)
        stop_time = int(round(time() * 1000))
        timedfun.time = stop_time-start_time
        return res
    return timedfun

@my_timer
def Fibonacci_ric(n):
    if n == 0 or n == 1:
        return n
    return Fibonacci_ric(n-2)+Fibonacci_ric(n-1)

@my_timer
def Fibonacci(n):
    f0 = 0
    f1 = 1
    for _ in range(n-1):
      f0, f1 = f1, f0+f1
    return f1
````

### Metodo \_\_Call\_\_: come modificarlo e utilizzarlo

Ipotizziamo di dover implementare un semplice generatore pseudo-casuale. Il generatore produce una sequenza di numeri $<R_n>$ nell'intervallo \[0,1), così definiti:

$$\begin{eqnarray*}
X_n&=&a\cdot X_{n-1} \mathrm{mod}\ m\\
R_n&=&\frac{X_n}{m}
\end{eqnarray*}
$$
con $n=1,2,\ldots$, dove $a, m$ e $X_0$ sono opportunamente scelti

````python
class myrand:
    '''generatore pseudo-casuale'''
    def __init__(self):
        from time import time
        self._a = 16807
        self._m = (1 << 31) - 1 #m = 2^31 -1
        self._d = 1.0 / self._m
        self._x = int(time())%self._m #x_0 "casuale", dipende dal tempo
        if self._x ==0:
            self._x = 1
        for i in range(10000):
            #facciamo un po' di giri a vuoto per rendere il numero indipendente dal seme iniziale
            
                self.__call__()
                
    def __call__(self):
        self._x =(self._a*self._x)%self._m
        return self._x*self._d
            
````

````python
r = myrand()

for _ in range(10):
    print(r()) #r() è zucchero sintattico per r.__call__()
````

L'esempio mosta come si può utilizzare il metodo \_\_call\_\_ per chiamarlo con lo stesso nome della classe; questo ne rende l'uso di comprensione immediata ed è un modo per realizzare "funzioni con stato".

## <span style="color:red">Class (as) decorators</span>

Come nell'esempio precedente, il metodo \_\_call\_\_ definito all'interno di una classe ci permette di rendere gli oggetti "richiamabili". In quanto, appunto, sono richiamabili, possiamo utilizzarli come decorator nello stesso modo nel quale lo abbiamo fatto con le funzioni.

Per esempio, per descrivere questi decoratori consideriamo il problema di generare numeri pseudo-casuali la cui funzione di densità probabilistica sia un'esponenziale negativa:

$$
f(x;\lambda) = \left\{\begin{matrix}
\lambda e^{-\lambda x} & x \ge 0, \\
0 & x < 0.
\end{matrix}\right.
$$

dove $\lambda>0$ è un parametro che può essere interpretato come *intensità* (di un fenomeno) o *frequenza* (di un evento).


La funzione di distribuzione probabilistica dell'esponensiale negativa di parametro $\lambda$ è $F(x;\lambda)=1-e^{-\lambda x}$ e si può dimostrare che, se $\{r_n\}$ è una sequenza di numeri casuali uniformemente distribuiti in $[0,1)$ allora $\{F^{-1}(r_n)\}$ è una sequenza distribuita con esponenziale negativa. Risulta quindi:
$$F^{-1}(r) = -\frac{\ln(1-r)}{\lambda}$$

````python
from math import log
class negexp:
    '''classe che implementa un generatore uniforme che produce
    numeri distribuiti con esponenziale negativo '''
    
    def __init__(self,LAMBDA): 
        #lambda (minuscolo) è una parola riservata di python
        self._lambda = LAMBDA
        
    def __call__(self,fn):
        '''fn è un generatore uniforme'''
        def decorated():
            r = fn()
            return -log(1.0-r)/self._lambda
        return decorated
````

````python
from random import random
````

Ci sono due metodi differenti per decorare le classi:
   
   * 1) decorare uno, più o tutti i suoi metodi
   * 2) decorare l'intera classe
   
#### Decoratori built-in per metodi

   1) <span style="color:red">@property</span> : viene utilizzato per specializzare l'accesso in lettura e scrittura a particolari attributi della classe
   
   2) <span style="color:red">@classmethod</span> : Un metodo decorato in questo modo non viene invocato sulle istanze di una classe. Viene invece utilizzato per creare istanze "specializzate" di oggetti della classe
   
   3) <span style="color:red">@staticmethod</span> : Un metodo statico può essere chiamato sia utilizzando la classe, sia utilizzando un oggetto della classe in quanto esso è legato alla classe per il fatto di fare parte del suo namespace.

````python
from math import log
class negexp:
    '''implementa un generatore uniforme che genera numeri
    distribuiti con esponenziale negativo'''
    
    def __init__(self,LAMBDA):
        self._lambda = LAMBDA
        
    @classmethod
    def unitfreq(cls):
        return cls(1.0)
    
    def __call__(self,fn):
        #self.rand = fn
        def decorated():
            #r = self.rand()
            r = fn()
            return -log(1.0-r)/self._lambda
        return decorated
````

````python
from random import random
@negexp.unitfreq()
def schedule():
    return random()
````

a.\_\_getattribute\_\_('w') non è la stessa cosa di scrivere a.w

Nel caso di *a.w*, l'interprete quando vede che \_\_getattribute\_\_ sollelva un eccezione, chiama \_\_getattr\_\_

````python
class prova2:
    '''miglioramento di prova1 gestendo il problema esposto'''
    
    y = 'un altro attributo della classe'
    
    def __init__(self, x = 0):
        '''ogni oggetto ha un solo attributo dinamico (in questo caso e in quello precedente)'''
        self.x = x
    
    def __getattribute__(self,attr):
        '''effettuiamo il miglioramento'''
        
        try:
            print(f"Look up dell'attributo {attr}")
            return object.__getattribute__(self, attr)
        except AttributeError:
            try:
                return object.__getattr__(self, attr)
            except AttributeError:
                print(f"L'attributo {attr} e il metodo __getattr__ non sono definiti")
                return None
            
    def __getattr(self, attr):
        '''gestisce l'eccezione AttributeError '''
        print(f"se l'attributo {attr} non è definito viene chiamata __getattr__()")
        return None
````

### Decoratore *@property*

Un modo Pythonico di forzare alcuni vincoli sul valore di un certo attributo consiste nel definire tale attributo come ***property***. Questo obiettivo può essere ottenuto usando il decorator *property* associato ai metodi getter e setter.

````python
class CC:
    
    '''classe per la gestine di un conto corrente:: v2.01
    evidenziamo scritture alterntive per specificare che un dato
    attributo è in realtà una proprietà'''
    
    def __init__(self, deposito_iniziale =0):
        print("Apertura conto con {0:.2f} Euro".format(deposito_iniziale))
        self.saldo = deposito_iniziale
        
    def prelievo(self, importo=0):
        #se importo == 0 equivale a una lettura del saldo
        if importo > 0 and self._saldo-importo>=0:
            self._saldo -=importo
        return self._saldo
    
    def __modifica_saldo(self, importo):
        if importo >= 0:
            self._saldo = importo
            
    def versamento(self, importo):
        if importo >0:
            self._saldo +=  importo
    
    saldo = property(prelievo, __modifica_saldo) #getter e setter
    
    #Equivalente
    
    #saldo = property(fget=prelievo, fset=__modifica_saldo)
    #e quindi anche a
    #saldo = property(fset=__modifica_saldo,fget = prelievo)
    
    #un altro modo
    #saldo = property()
    #saldo = saldo.getter(prelievo)
    #saldo = saldo.setter(__modifica_saldo)
````

````python
class CC:
    '''
    Classe per la gestione di un contocorrente:: v2.1
    riorganizza i metodi allo scopo di usare i decoratori
    '''
    
    def __init__(self, deposito_iniziale = 0):
        print("apertura conto con {0:.2f} Euro".format(deposito_iniziale))
        self._saldo = deposito_iniziale
        
    @property
    def saldo(self):
        return self._saldo
    
    @saldo.setter
    def saldo(self,value):
        if value > 0:
            self._saldo = value
            
    def versamento(self, importo):
        if importo > 0:
            self.saldo += importo
            
    def prelievo(self,importo):
        if importo > 0 and self.saldo>=importo:
            self.saldo -= importo
            
    def mostra_saldo(self):
        print("il saldo è di {0:.2f} euro".format(self._saldo))
````

### Descrittori

Un descrittore è una classe in cui è presente la definizione di almeno uno dei seguenti *magic methods*:

   1. **get***(self, instance, cls)
   2. ***set***(self, instance, value)
   3. ***delete***(self, instance)
   
I descrittori vengono utilizzati per controllare l'accesso agli attributi di altre classi. Sono una generalizzazione delle proprietà.

````python
class desc:
    
    def __init__(self, name = "attr", initval = None):
        self.name = name
        self.val = initval
        
    def __get__(self, inst, cls):
        '''
        in questo esempio inst e cls non sono utilizzati.
        - inst è l'oggetto in cui il descrittore è utilizzato
        - cls è il suo tipo (la sua classe)
        Il valore restituito potrebbe dipendere quindi da altri attributi
        definiti nell'oggetto in questione.
        '''
        print("__get__ called")
        return self.val
    
    def __set__(self, inst, val):
        '''
        anche qua inst non viene utilizzato
        '''
        
        print("__set__ called")
        if val < 0 or val > 1:
            raise ValueError(f"il valore di {self.name} dovrebbe essere contenuto nell'intervallo [0,1]")
        else:
            self.val = val
````

Questo esempio è troppo semplice, il fatto che il descrittore venga utilizzato come attributo di classe ha diversi vantaggi (ad esempio in termini di efficienza). E non pregiudica la possibilità di utilizzarlo per istanze diverse.


````python
class balance:
    '''Classe il cui uso inteso è di memorizzare il (e operare sul)
       saldo di molteplici conti correnti. Poiché implementa i metodi
       __get__, __set__ e __delete__, la classe è un descrittore.
       La classe utilizza un dizionario (scelta migliorabile) per 
       memorizzare il saldo di conti correnti diversi. Le chiavi di accesso
       sono proprio gli oggetti (della classe CC) che rappresentano
       i conti correnti aperti. Il riferimento a tali oggetti è il
       parametro instance dei vari metodi. Gli accessi in lettura e scrittura
       al descrittore (accessi che vengono effettuati effettuati nei metodi
       della classe CC) sono esprimibili in "pure python" nel modo seguente,
       in cui b è il balance e X e il CC (CC=type(X)):
           X.b (lettura)         --> type(X).__dict__['b'].__get(X,type(X))
           X.b = 100 (scrittura) --> type(X).__dict__['b'].__set(X,100)
    '''
    
    def __init__(self):
        self.default = 0
        self.balances = {}
        
    def __get__(self,instance,cls):
        '''Usando come chiave l'istanza (di CC) passata come secondo 
           parametro, il metodo recupera il valore del saldo associato
           dal dizionario self.balances
        '''    
        return self.balances.get(instance,self.default)
    
    def __set__(self, instance, value):
        '''Usando come chiave l'istanza (di CC) passata come secondo 
           parametro, il metodo "setta" il saldo corrispondente nel
           dizionario self.balances al valore value
        '''
        if value <-instance.credit:
            raise ValueError("superato il valore del fido")
        self.balances[instance] = value
    
    def __delete__(self,instance):
        ''' 
        Cancella la coppia istanza/valore dal dizionario self.balances
        '''
        del self.balances[instance]
    
````

````python
class CC:
    """
    Classe per la gestione di un c/c:: v3.0
    Utilizza un descrittore per la memorizzazione e la manipolazione dei saldi.
    Introduce l'utilizzo del fido
    """
    def __init__(self,initBalance=0,BankCredit=0):
        ''' 
        Apre un conto con assegnato saldo iniziale e dato fido bancario
        '''
        self.credit = BankCredit
        self.b = initBalance
        
    def deposit(self,value):
        ''' 
        Deposita un valore positivo sul conto
        '''
        if value>0:
            self.b += value
            
    def withdraw(self, value):
        '''
        Ritira una data soma dal conto (i controlli sull'eventuale
        superamento del fido sono effettuati dal descrittore)
        '''
        self.b -= value
        
    def getbalance(self):
        '''
        Stampa il saldo
        '''
        print(f"Current balance is {self.b}")
    
    b = balance() # Il saldo è un descrittore
````

### A cosa servono?

Le classi astratte vengono solitamente utilizzateper definire funzionalità di default per le sue sottoclassi. Questo concetto è simile a quello di ***interfaccia (java)***, ma in realtà non sono la stessa cosa in quanto, a differenza delle interfacce, le classi astratte possono essere parzialmente implementate (come nell'esempio 5 blocchi sopra a questo blocco).

Ad esempio, nel caso in cui due sottoclassi hanno metodi o attributi comuni, essi verranno definiti astratti nella classe padre.

````python
from abc import ABC, abstractmethod
import math
import pygame
# conda update --all
# pip install pygame
black = (0,0,0)
white = (255,255,255)
red = (255,0,0)
green = (0,255,0)
blue = (0,0,255)

class mydisplay:
    '''
    classe che definisce una schermata grafica utilizzata per
    disegnare "singoli" oggetti geometrici
    '''
    
    def __init__(self, width=800,height=600):
        pygame.display.init()
        self.width = width
        self.height = height
        self.display = pygame.display.set_mode((width,height))
        
    @staticmethod
    def close():
        pygame.quit()
        
    def clean(self):
        self.display.fill(white)
        
    def drawline(self, p, q, color = black):
        self.clean()
        pygame.draw.line(self.display,color,(p[0], self.height-p[1]),(q[0],self.height-q[1]), 3)
        pygame.display.update()
    
    def drawcircle(self, c, r , color=black):
        self.clean()
        pygame.draw.circle(self.display, color, (c[0],self.height-c[1]),r)
        pygame.display.update()
        
    def drawrect(self, bl, w, h, color=black):
        self.clean()
        pygame.draw.rect(self.display, color, pygame.Rect(bl[0], self.height-bl[1],w,h))
        pygame.display.update()
        
class oggetto_geometrico(ABC):
    '''
    La classe definisce un oggetto astratto generico.
    Ovviamente le classi forniranno le implementazioni dei vari 
    oggetti con diverse funzioni
    '''
    
    def __init__(self, display, color = black):
        self.display = display
        self.color = color
        
    @staticmethod
    def ruotaV(x,y, theta):
        xp = x*math.cos(theta)-y*math.sin(theta)
        yp = y*math.sin(theta)+y*math.cos(theta)
        return xp,yp
    
    @abstractmethod
    def sposta(self,dx,dy):
        pass
    
    @abstractmethod
    def ruota(self, theta):
        pass
````

````python
class line(oggetto_geometrico):
    
    def __init__(self,p,q,display,color = black):
        super().__init__(display,color)
        self.p = p
        self.q = q
        self.display.drawline(p,q,color)
        
    def sposta(self, dx, dy, disegna=True):
        from operator import add
        self.p = tuple(map(add,self.p,(dx,dy)))
        self.q = tuple(map(add,self.q,(dx,dy)))
        disegna and self.display.drawline(self.p,self.q,self.color)
        
    def ruota(self,theta):
        cx = (self.p[0]+self.q[0])/2
        cy = (self.p[1]+self.q[1])/2
        self.sposta(-cx,-cy, disegna=False)
        self.p = self.ruotaV(*self.p,theta)
        self.q = self.ruotaV(*self.q,theta)
        self.sposta(cx,cy)
````

## Metaclassi

Una metaclasse mette a disposizione meccanismi grazie ai quali è possibile personalizzare l'instanziazione di una classe.

Una classe è un'istanza di type ed è istanza di se stessa. Quindi Type è una ***metaclasse*** e sono metaclassi tutte le classi che ereditano da type.

````python
class metaclasse(type):
    def __init__(cls,name, bases, dictionary):
        '''Inizializzazione che esegue solo delle stampe
        ci permettono di caire quando viene eseguito il codice
        e i parametri passati'''
        print('cls',cls)
        print('name',name)
        print('bases',bases)
        print('dictionary',dictionary)
    foo = 0
````

Possiamo fare in modo che il tipo di una classe sia una metaclasse diversa da type utilizzando la seguente sintassi

```python
    class NOMECLASSE(elenco_superclassi, metaclass=nome_metaclasse):
        corpo_metaclasse
```




````python
class S:
    foo = 10
    
class X(S,metaclass = metaclasse):
    pass
````

Quindi cosa è successo?

Quando viene creata una classe (in questo caso X) il cui tipo è una metaclasse diversa da type:

   1. vengono chiamati i metodi \_\_new\_\_ e \_\_init\_\_ della metaclasse
   2. poichè (ovviamente) valgono i meccanismi di ereditarietà e la metaclasse eredita da type, se la metaclasse non definisce uno di tali metodi allora viene chiamato il metodo corrispondente di type
   
Nel nostro caso ***metaclasse*** definisce \_\_init\_\_ ma non \_\_new\_\_, quindi è stato prima chiamato il metodo \_\_new\_\_ di tyoe e poi il metodo \_\_init\_\_ di ***metaclasse***

Potremmo riscrivere il metodo \_\_new\_\_

````python
class metaclasse(type):
    def __new__(cls,name,bases,dictionary):
        '''Intercettiamo __new__ e chiamiamo esplicitamente
        la new di type'''
        print('cls',cls)
        print('name',name)
        print('bases',bases)
        print('dictionary',dictionary)
        return type.__new__(cls,name,bases,dictionary)
    foo = 10
    
class X(S,metaclass=metaclasse):
    '''X eredita da tuple ma il suo tipo è "metaclasse"'''
    foo = 5
    pass
````

````python
x = X()
# Quali saranno i valori stampati dai seguenti comandi?
print(x.foo)
print(type(x).foo)
print(type(X).foo)
print(type(type(x)).foo)
````

### Schema algoritmico map-reduce

Si tratta di un procedimento di calcolo tipico dei linguaggi funzionali e presente in Python.

Un primo semplice esempio che illustra la struttura del paradigma, riguarda il *calcolo della somma dei quadrati dei primi n numeri naturali*

````python
#versione C style
def sum_of_squares_C(n):
    s = 0
    for i in range(1,n+1):
        s+=i*i
    return s
````

````python
sum_of_squares_C(10)
````

````python
#versione Python style con map-reduce
def sum_of_squares_P(n):
    s = [i*i for i in range(1,n+1)] #map step
    return sum(s) #reduce step
````

````python
sum_of_squares_P(10)
````

La seconda versione illustra due step dello schema map-reduce:
   
   1. ***map*** : una stessa operazione viene applicata singolarmente ad ogni dato che compone la collezione di input
   
   2. ***reduce***: un'operazione (tipicamente binaria) viene applicata ai dati prodotti in output da map in modo da produrre un unico risultato in uscita
   
La versione in Map-Reduce presenta un netto svantaggio nel richiedere memoria $O(n)$, al constrario della prima che usa una quantità costante di spazio. Nonostante ciò, la versione Map-Reduce, in particolare al primo passo, può essere parallelizzata a livello computazionale. Implementazioni di questo schema si trovano non a caso in sistemi generali di calcolo distribuito come *Spark* e *Hadoop* in Apache. In contesti reali, lo schema può essere più complesso e può precedere un passo intermedio di *shuffling*.

#### Supporto di "base" per map-reduce in Python

````python
def sum_of_squares_P(n):
    from functools import reduce
    s = map(lambda i: i*i, range(1,n+1))
    return reduce (lambda x,y : x+y,s)
````

````python
sum_of_squares_P(10)
````

In questa seconda versione Map-Reduce, il passo map viene eseguito utilizzando l'omonima funzione di ordine superiore.

   1. ***map*** riceve in input una funzione e un iterabile e applica la funzione a tutti gli elementi dell'iterabile restituendo, in output, un secondo iterabile con i valori calcolati
   
   2. ***reduce***, una volta ricevuta in input una funzione binaria $f$ e un iterabile $I$, esegue i seguenti $|I| - 1$ passi:
   
       2.1 al primo passo applica $f$ ai primi due elementi di $I$
       
       2.2 al passo $k, k = 2, ..., |I| - 1 $, applica $f$ al risultato ottenuto al passo $k - 1$ e al $k + 1$-esimo elemento di $I$

````python
#esempio 1: calcolo fattoriale con reduce
from functools import reduce
reduce(lambda x,y: x*y, range(1,6))
````

````python
# Ordinamento (shuffling) usando memoria centrale "controllata"
# See https://stackoverflow.com/questions/14465154/sorting-text-file-by-using-python/16954837#16954837

import sys
from functools import partial
from heapq import merge
from tempfile import TemporaryFile

# defininiamo il criterio di sorting
def sortkey(line):
    return line.split('\t')[0]

with open('testo.map') as infile:
    # sort lines in small batches, write intermediate results to temporary files
    sorted_files = []
    nbytes = 1 << 16 # Carichiamo 64KB per volta 
    for lines in iter(partial(infile.readlines, nbytes), []):
        lines.sort(key=sortkey)     # ordiniamo in memoria il batch (di al più nbytes) attuale
        f = TemporaryFile("w+")
        f.writelines(lines)
        f.seek(0) # rewind
        sorted_files.append(f)

# fusione e scrittura dei risultati
with open('testo.map.sorted','w') as sortedmap:
    sortedmap.writelines(merge(*sorted_files, key=sortkey))

# clean up
for f in sorted_files:
    f.close() # chiudiamo e cancelliamo i file temporanei
````
